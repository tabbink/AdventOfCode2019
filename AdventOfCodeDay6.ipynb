```python
#Circle around planets

#We need to calculate the amount of direct and indirect orbits around a planet
#recursively?: If the parent directly orbits another planet, the child indirectly orbits the planet
# for indirect: work from the front to the back.
# step 1: find all planets that orbit nothing
# step 2: find all planets that orbit the "static" planets found in step 1
# step 3: count how many direct orbits the "static" planet had (first iteration this will be 0), these are the amount of indirect orbits for the orbitting planets
# step 4: remove the "static" planets
# step 5: repeat from step 1 till there are no orbitting planets left


def TotalAmountofOrbits(map_data):
    def SplitOrbit(map_data):
        map_data_splitted = [i.split(')') for i in map_data]
        return [item[0] for item in map_data_splitted], [item[1] for item in map_data_splitted]

    def TotalDirectOrbits(map_data, unique_planet_list):
        for i in map_data['orbit_planet']:
            unique_planet_list.loc[unique_planet_list['planet'] == i, 'CountDirectOrbits']  += 1
        return

    map_data['static_planet'], map_data['orbit_planet'] = SplitOrbit(map_data['raw_data'])
    unique_planet_list = pd.DataFrame({'planet':map_data['orbit_planet'].unique()}) # This totally ignores all static planets
    unique_planet_list = unique_planet_list.append(pd.DataFrame({'planet':map_data['static_planet'].unique()}), ignore_index=True).drop_duplicates('planet') #this adds the static planets
    unique_planet_list['CountDirectOrbits'], unique_planet_list['CountIndirectOrbits'], unique_planet_list['Path'] = 0, 0, ''

    TotalDirect = TotalDirectOrbits(map_data, unique_planet_list)
    return unique_planet_list
```

```python
def calculate_indirect(map_data, unique_planet_list, parent):
    parent_data = unique_planet_list[unique_planet_list['planet'] == parent]
    for i in range(0, len(map_data)):
        if map_data['static_planet'].loc[i] == parent:
            unique_planet_list.loc[unique_planet_list['planet'] == map_data['orbit_planet'].loc[i], 'CountIndirectOrbits'] = parent_data['CountDirectOrbits'].values[0] + parent_data['CountIndirectOrbits'].values[0]
            unique_planet_list.loc[unique_planet_list['planet'] == map_data['orbit_planet'].loc[i], 'Path'] = parent_data['Path'].values[0] + parent + ','

    children_data = map_data[map_data['static_planet'] == parent]
    for i in range(0, len(children_data)):
        test = children_data['orbit_planet'].iloc[i]
        calculate_indirect(map_data, unique_planet_list, test)
    
    return
```
