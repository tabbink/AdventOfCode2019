```python
def IntCode_Computer(IntCodeInput, Opcodelocation, inputvalue, iterator = 0, outputvalue = 0):
    IntCode = IntCodeInput[:]
    def DetermineOpcodeNumber(number):
        return int(str(number)[-2:])
    
    def DetermineParameterArrayLocations(Intcode, Opcodelocation):
        def DetermineParameterMode(number):
            firstparameter = int(str(number // 100)[-1:])
            secondparameter = int(str(number // 1000)[-1:])
            thirdparameter = int(str(number // 10000)[-1:])
            return firstparameter, secondparameter, thirdparameter

        def DetermineParameterLocations(Intcode, Opcodelocation, parameters):
            # als parameter 0 is, dan verwijzing. Als parameter 1, dan eigen locatie
            firstparameterlocation = Opcodelocation + 1 if parameters[0] == 1 else Intcode[Opcodelocation + 1]
            try:
                secondparameterlocation = Opcodelocation + 2 if parameters[1] == 1 else Intcode[Opcodelocation + 2]
            except IndexError:
                secondparameterlocation = 0
            try:
                thirdparameterlocation = Opcodelocation + 3 if parameters[2] == 1 else Intcode[Opcodelocation + 3]
            except IndexError:
                thirdparameterlocation = 0
            return firstparameterlocation, secondparameterlocation, thirdparameterlocation

        parameters = DetermineParameterMode(Intcode[Opcodelocation])
        return DetermineParameterLocations(Intcode, Opcodelocation, parameters)

    def RunOpcode(Intcode, OpcodeNumber, parameterlocations, inputvalue, iterator, outputvalue):
        def Opcode1(Intcode, parameterlocations, inputvalue, iterator, outputvalue):
            Intcode[parameterlocations[2]] = Intcode[parameterlocations[0]] + Intcode[parameterlocations[1]]
            return IntCode_Computer(Intcode, Opcodelocation + 4, inputvalue, iterator, outputvalue)
        def Opcode2(Intcode, parameterlocations, inputvalue, iterator, outputvalue):
            Intcode[parameterlocations[2]] = Intcode[parameterlocations[0]] * Intcode[parameterlocations[1]]
            return IntCode_Computer(Intcode, Opcodelocation + 4, inputvalue, iterator, outputvalue)
        def Opcode3(Intcode, parameterlocations, inputvalue, iterator, outputvalue):
            if iterator == 0:
                Intcode[parameterlocations[0]] = inputvalue
                iterator = 1
                return IntCode_Computer(Intcode, Opcodelocation + 2, inputvalue, iterator, outputvalue)
            else:
                Intcode[parameterlocations[0]] = outputvalue 
                #iterator = 0
                return IntCode_Computer(Intcode, Opcodelocation + 2, inputvalue, iterator, outputvalue)
        def Opcode4(Intcode, parameterlocations, inputvalue, iterator, outputvalue):
            outputvalue = Intcode[parameterlocations[0]]
            return [Intcode, Opcodelocation + 2, inputvalue, iterator, outputvalue]
        def Opcode5(Intcode, parameterlocations, inputvalue, iterator, outputvalue):
            if Intcode[parameterlocations[0]] != 0:
                return IntCode_Computer(Intcode, Intcode[parameterlocations[1]], inputvalue, iterator, outputvalue)
            else:
                return IntCode_Computer(Intcode, Opcodelocation + 3, inputvalue, iterator, outputvalue)
        def Opcode6(Intcode, parameterlocations, inputvalue, iterator, outputvalue):
            if Intcode[parameterlocations[0]] == 0:
                return IntCode_Computer(Intcode, Intcode[parameterlocations[1]], inputvalue, iterator, outputvalue)
            else:
                return IntCode_Computer(Intcode, Opcodelocation + 3, inputvalue, iterator, outputvalue)
        def Opcode7(Intcode, parameterlocations, inputvalue, iterator, outputvalue):
            if Intcode[parameterlocations[0]] < Intcode[parameterlocations[1]]:
                Intcode[parameterlocations[2]] = 1
            else:
                Intcode[parameterlocations[2]] = 0
            return IntCode_Computer(Intcode, Opcodelocation + 4, inputvalue, iterator, outputvalue)
        def Opcode8(Intcode, parameterlocations, inputvalue, iterator, outputvalue):
            if Intcode[parameterlocations[0]] == Intcode[parameterlocations[1]]:
                Intcode[parameterlocations[2]] = 1
            else:
                Intcode[parameterlocations[2]] = 0
            return IntCode_Computer(Intcode, Opcodelocation + 4, inputvalue, iterator, outputvalue)

        if OpcodeNumber == 1:
            return Opcode1(Intcode, parameterlocations, inputvalue, iterator, outputvalue)
        elif OpcodeNumber == 2:
            return Opcode2(Intcode, parameterlocations, inputvalue, iterator, outputvalue)
        elif OpcodeNumber == 3:
            return Opcode3(Intcode, parameterlocations, inputvalue, iterator, outputvalue)
        elif OpcodeNumber == 4:
            return Opcode4(Intcode, parameterlocations, inputvalue, iterator, outputvalue)
        elif OpcodeNumber == 5:
            return Opcode5(Intcode, parameterlocations, inputvalue, iterator, outputvalue)
        elif OpcodeNumber == 6:
            return Opcode6(Intcode, parameterlocations, inputvalue, iterator, outputvalue)
        elif OpcodeNumber == 7:
            return Opcode7(Intcode, parameterlocations, inputvalue, iterator, outputvalue)
        elif OpcodeNumber == 8:
            return Opcode8(Intcode, parameterlocations, inputvalue, iterator, outputvalue)

    if IntCode[Opcodelocation] == 99:
        return [IntCodeInput, -999, inputvalue, iterator, outputvalue]
    else:
        return RunOpcode(IntCode, DetermineOpcodeNumber(IntCode[Opcodelocation]), DetermineParameterArrayLocations(IntCode, Opcodelocation), inputvalue, iterator, outputvalue)
```
